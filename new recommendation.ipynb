# TF-IDF Vectorizer 

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
user_description = "User's description of the place"
```

```python
cv = CountVectorizer()
cv_matrix = cv.fit_transform(data['description'] + ' ' + user_description)
```

```python
cosine_sim = cosine_similarity(cv_matrix)
cosine_sim_df = pd.DataFrame(cosine_sim, index=data['name'], columns=data['name'])
```

```python
# Recommendation function
def tourism_recommendations(user_description, similarity_data=cosine_sim_df, items=data[['name','category','description','city']], k=5):
    # Preprocessing and feature extraction
    cv_user = CountVectorizer(vocabulary=cv.vocabulary_)
    cv_user_matrix = cv_user.transform([user_description])

    # Calculate cosine similarity between user description and place descriptions
    cosine_sim_user = cosine_similarity(cv_user_matrix, cv_matrix)
    cosine_sim_user = cosine_sim_user.flatten()

    # Get top-k similar places
    indices = cosine_sim_user.argsort()[:-k-1:-1]
    closest = items.loc[indices, 'name']

    return pd.DataFrame(closest).merge(items).head(k)

# Example usage
user_description = "want to visit historical places wants to see Tanks and guns"

# Preprocessing and feature extraction
cv = CountVectorizer()
cv_matrix = cv.fit_transform(data['description'])

# Cosine similarity calculation
cosine_sim = cosine_similarity(cv_matrix)
cosine_sim_df = pd.DataFrame(cosine_sim, index=data['name'], columns=data['name'])

# Call the recommendation function
recommendations = tourism_recommendations(user_description)
print(recommendations)
```
